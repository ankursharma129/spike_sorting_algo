```python
import numpy as np
import scipy.signal as signal
import pylab as pl

# Calculate the coefficients for a pure fixed point
# integer filter

# sampling rate
fs = 20000

# cutoffs
f1 = 300
f2 = 6500

# scaling factor in bits
q = 14
# scaling factor as facor...
scaling_factor = 2**q

# let's generate a sequence of 2nd order IIR filters
sos = signal.butter(3,[f1/fs*2,f2/fs*2],'band',output='sos')
print("SOS coefficients before scaling", sos)

sos = np.round(sos * scaling_factor)
print("SOS coefficients after scaling", repr(sos))
```
